```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.array([1, 1])
y = np.array([2, 0])
z = np.array([0, 2])
```

```python
plt.figure()
plt.plot(x[0], x[1], 'o')
plt.plot(y[0], y[1], 'o')
plt.plot(z[0], z[1], 'o')
plt.grid()
```

```python
xyz = x+y+z
```

```python
plt.figure()
plt.plot(xyz[0], xyz[1], 'o')
plt.grid()
```

```python
dot = np.dot(x, y)
dot
```

```python
x_len = (x[0] ** 2 + x[1] ** 2) ** 0.5
x_len
```

```python
y_len = (y[0] ** 2 + y[1] ** 2) ** 0.5
y_len
```

```python
cos_ang = dot/(x_len * y_len)
```

```python
math.degrees(math.acos(cos_ang))
```
